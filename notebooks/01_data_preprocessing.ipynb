```python
from Korpora import Korpora
import pandas as pd
```

```python
corpus = Korpora.load("korean_hate_speech")

train_df = pd.DataFrame(corpus.train)
dev_df = pd.DataFrame(corpus.dev)

print(train_df.shape)
print(dev_df.shape)
train_df.head()
```

```python
train_df.columns
```

```python
train_df["hate"].value_counts()
```

```python
def map_label(x):
    if x in ["hate", "offensive"]:
        return 1
    return 0

train_df = train_df.rename(columns={"text": "comments"})
dev_df = dev_df.rename(columns={"text": "comments"})

train_df["label"] = train_df["hate"].apply(map_label)
dev_df["label"] = dev_df["hate"].apply(map_label)
train_df[["comments", "hate", "label"]].head()
```

```python
train_df = train_df.dropna(subset=["comments"])
train_df["comments"] = train_df["comments"].astype(str).str.strip()
train_df = train_df[train_df["comments"] != ""]
train_df = train_df[train_df["comments"].str.len() >= 3]
train_df = train_df.drop_duplicates(subset=["comments"])

train_df.shape
```
